```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics.pairwise import cosine_similarity

def tfidf_vec():
    data = fetch_20newsgroups(subset='train')
    documents = data.data
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)
    return vectorizer,tfidf_matrix, documents

def doc_sim(input, vectorizer, tfidf_matrix, docs):
    input_vector = vectorizer.transform([input])
    similarities = cosine_similarity(input_vector, tfidf_matrix).flatten()
    ranked_documents = sorted(enumerate(similarities), key=lambda x: x[1], reverse=True)
    doc_indices, scores = zip(*ranked_documents)
    ranked_docs = [docs[i] for i in doc_indices]
    return ranked_docs, scores

input = "Good Morning"
vectorizer, tfidf_matrix, docs = tfidf_vec()
similar_documents, similarity_scores = doc_sim(input, vectorizer, tfidf_matrix, docs)
print("Input Document :-", input)
print(input)
print("\n\n\n")
print("Top 5  matching documents are as follows :-")
for i in range(5):
   print("-->",end='')
   print(f"Similarity Score:",end='')
   print(similarity_scores[i])
   print(similar_documents[i])
   print()
```
