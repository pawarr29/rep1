```python
import numpy as np
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

```

Q1.

```python
np.random.seed(10) # Setting the seed of 10 for drawing the randomnumbers of size 100
random_numbers = np.random.randint(0, 11, size=100) # drawing 100 randim numbers from 0 to 11
```

```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate() # crating new one
rdd_numbers = sc.parallelize(random_numbers) # parallelizing the data and creatinf rdd
```

```python
number_frequencies = rdd_numbers.map(lambda number: (number, 1)).reduceByKey(lambda a, b: a + b)
frequencies = number_frequencies.collect()
for number, frequency in sorted(frequencies):
    print(f"Number: {number}, Frequency: {frequency}")
```

Q2.

```python
sc = SparkContext.getOrCreate()
rdd = sc.textFile("/content/text8") # storing ion new rdd
```

```python
words = rdd.flatMap(lambda line: line.split(" ")) # Splitting the words through space
```

```python
word_frequencies = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
# mapping a word a frequncy of 1 and reducing the list by adding the frequency of the same words
```

```python
words_with_a = word_frequencies.filter(lambda word_freq: 'a' in word_freq[0])
# filtering the words with 'a'
```

```python
for word, count in words_with_a.collect(): # collecting and priting the same
    print(f"{word}: {count}")
```

```python
# Creating a new rdd by input of the iris dataset
spark = SparkSession.builder.appName("IrisDataFrame").getOrCreate()
```

```python
# reading json file of iris dataset
df = spark.read.json("/content/iris.json")
```

```python
# Finding preasons correlation.
pearson_correlation = df.stat.corr("petalLength", "petalWidth")
print(f"Pearson Correlation Coefficient between petalLength and petalWidth: {pearson_correlation}")
```

```python
filtered_df = df.filter(col("petalLength") >= 1.4).select("sepalLength", "sepalWidth", "species")
filtered_df.show() # action command called
```
